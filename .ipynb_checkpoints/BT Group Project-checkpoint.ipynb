```python
card = card.rename(columns={'PAY_0': 'PAY_1','default payment next month': 'defPay'})
```

```python
# Data Cleaning: for education
fil = (card.EDUCATION == 5) | (card.EDUCATION == 6) | (card.EDUCATION == 0)
card.loc[fil, 'EDUCATION'] = 4
card.EDUCATION.value_counts()
```

```python
# Data Cleaning: for marriage
card.loc[card.MARRIAGE == 0, 'MARRIAGE'] = 3
card.MARRIAGE.value_counts()
```

```python
# Data Cleaning: for PAY_1 to PAY_6
def histogram(data, variables, n_rows, n_cols, n_bins):
    fig=plt.figure()
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        data[var_name].hist(bins=n_bins,ax=ax)
        ax.set_title(var_name)
    fig.tight_layout()  # Improves appearance a bit.
    plt.show()

p = (card.PAY_1 == -2) | (card.PAY_1 == -1) | (card.PAY_1 == 0)
card.loc[p, 'PAY_1'] = 0
p = (card.PAY_2 == -2) | (card.PAY_2 == -1) | (card.PAY_2 == 0)
card.loc[p, 'PAY_2'] = 0
p = (card.PAY_3 == -2) | (card.PAY_3 == -1) | (card.PAY_3 == 0)
card.loc[p, 'PAY_3'] = 0
p = (card.PAY_4 == -2) | (card.PAY_4 == -1) | (card.PAY_4 == 0)
card.loc[p, 'PAY_4'] = 0
p = (card.PAY_5 == -2) | (card.PAY_5 == -1) | (card.PAY_5 == 0)
card.loc[p, 'PAY_5'] = 0
p = (card.PAY_6 == -2) | (card.PAY_6 == -1) | (card.PAY_6 == 0)
card.loc[p, 'PAY_6'] = 0
late = card[['PAY_1','PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']]
```

```python
# to check correlation
plt.figure(figsize = (6,4))
sns.heatmap(card.corr(), cmap ='Blues', annot = False)
```

```python
#Quality correlation matrix
k = 24 #number of varaibles for heatmap
cols = card.corr().nlargest(k, 'defPay')['defPay'].index
cm = card[cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(cm, annot=True, cmap = 'viridis')
```

```python
#To check Outliers
l = card.columns.values
number_of_columns = 24
number_of_rows = len(l) -1/ number_of_columns
plt.figure(figsize=(number_of_columns,5*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows + 1,number_of_columns,i+1)
    sns.set_style('whitegrid')
    sns.boxplot(card[l[i]],color='green',orient='v')
    plt.tight_layout()
```

```python
#To check distribution-Skewness
plt.figure(figsize=(2*number_of_columns,5*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows + 1,number_of_columns,i+1)
    sns.distplot(card[l[i]],kde=True) 
```
