```python
#for rich
array = train_rich.values
X = array[:,0:23]
Y = array[:,23]

seed = 7

models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('LinearSVC', LinearSVC(random_state = 0)))
models.append(('SVM - rbf', SVC()))
#models.append(('RandomForest', RandomForestClassifier()))
models.append(('SVM - poly degree 8', SVC(kernel = 'poly', degree=8)))
models.append(('SVM - poly degree 3', SVC(kernel = 'poly', degree=3)))
models.append(('SVM - poly degree 2', SVC(kernel = 'poly', degree=2)))

#worthy to note is that when the degree is 4-9 for poly svm, the accuracy is the same as that for degree 8
#models.append(('SVM', SVC(kernel = 'linear')))
#models.append(('SVM', SVC(kernel = 'poly', degree=4)))


results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
    
```

```python
fig = plt.figure(figsize = (20, 12))
fig.suptitle('Algorithm Comparison for the rich')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
